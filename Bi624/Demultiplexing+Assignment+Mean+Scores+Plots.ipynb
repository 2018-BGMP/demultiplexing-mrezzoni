## Biological Read Mean Scores

```python
###R1

%matplotlib inline
import matplotlib.pyplot as plt

R1 = "/Users/evangelinarivas/shell/demultiplexing/1294_S1_L008_R1_001_means.txt"


x = []
y = []

with open(R1, "r") as fh:
    for line in fh:
        line = line.strip("\n").split("\t")
        y.append(float(line[1]))

plt.figure(figsize=(10,10))
plt.plot(y)

plt.title("Forward Read (R1) Mean Illumina Sequencing Quality Scores", fontsize=15)
plt.xlabel("Base Number", fontsize=12)
plt.ylabel("Mean Quality Score", fontsize=12)

plt.show()    
```
